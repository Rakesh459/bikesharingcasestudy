### Adding Dummy variables for categories

```python
import calendar
bikeData['mnth'] = bikeData['mnth'].apply(lambda x: calendar.month_abbr[x])
bikeData.season = bikeData.season.map({1: 'Spring',2:'Summer',3:'Fall',4:'Winter'})
## mapping weathersit
bikeData.weathersit = bikeData.weathersit.map({1:'Clear',2:'Mist & Cloudy', 
                                             3:'Light Snow & Rain',4:'Heavy Snow & Rain'})
bikeData.weekday = bikeData.weekday.map({0:"Sunday",1:"Monday",2:"Tuesday",3:"Wednesday",4:"Thrusday",5:"Friday",6:"Saturday"})

```

```python
bike_df = bikeData
```

```python
dummy = bike_df[['season','mnth','weekday','weathersit']]

dummy = pd.get_dummies(dummy,drop_first=True )
## adding dummy variables to original dataset
bike_df = pd.concat([dummy,bike_df],axis = 1)

# Now let's see the head of our dataframe.
bike_df.head()
```

```python
bike_df.drop(['season', 'mnth', 'weekday','weathersit','instant','dteday'], axis = 1, inplace = True)
```

```python
bike_df.shape
```

```python
bike_df_train, bike_df_test = train_test_split(bike_df,train_size=0.7,test_size = 0.3,random_state=100)
bike_df_train
```

### model building and evaluation

### Building a linear model

Fit a regression line through the training data using statsmodels. In statsmodels, you need to explicitly fit a constant using sm.add_constant(X) because if we don't perform this step, statsmodels fits a regression line passing through the origin, by default.

```python
#separating X and y
y_train = bike_df_train.pop('cnt')
X_train = bike_df_train
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
X_train.columns
```

```python
# Running RFE with the output number of the variable equal to 13
lm = LinearRegression()
lm.fit(X_train, y_train)

rfe = RFE(lm,n_features_to_select = 13)             # running RFE
rfe = rfe.fit(X_train, y_train)
```

```python
X_train.columns
```

```python
rfe.support_
```

```python
rfe.ranking_
```

```python
#Columns selected by RFE and their weights
list(zip(X_train.columns,rfe.support_,rfe.ranking_))
```

### Starting with all the columns selected by RFE
### Model 1

```python
#Print Columns selected by RFE. We will start with these columns for manual elimination
col = X_train.columns[rfe.support_]
col
```

```python
X_train.columns[~rfe.support_]
```

```python
# Creating X_train dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# Adding a constant variable 
 
X_train_rfe = sm.add_constant(X_train_rfe)
```

```python
# Create a first fitted model
lm = sm.OLS(y_train,X_train_rfe).fit()
```

```python
# Check the parameters obtained

lm.params
```

```python
# Print a summary of the linear regression model obtained
lm.summary()
```

```python
X_train_new = X_train_rfe.drop(["atemp"], axis = 1)
```

### Model 2

```python
# Adding a constant variable 
X_train_lm = sm.add_constant(X_train_new)
```

```python
lm = sm.OLS(y_train,X_train_lm).fit()   # Running the linear model
```

```python
lm.summary()
```

```python
# Calculate the VIFs for the new model
def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vff = []
    for i in range(df.shape[1]):
        try:
           val = variance_inflation_factor(df.values, i)  
           vff.append(val)
        except FloatingPointError:
           vff.append(np.inf)  
    vif['VIF'] = vff
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif

calculateVIF(X_train_new)
```

### Residual Analysis of the train data

### Inference

```python
param = pd.DataFrame(lm.params)
param.insert(0,'Variables',param.index)
param.rename(columns = {0:'Coefficient value'},inplace = True)
param['index'] = list(range(0,9))
param.set_index('index',inplace = True)
param.sort_values(by = 'Coefficient value',ascending = False,inplace = True)
param
```


### We can see that the equation of our best fitted line is:

#### cnt = 0.208847 + 0.488987 * temp + 0.234799 * yr + 0.048772 * seasonWinter -0.106449 * season Spring + -0.068388 * mnth_Jul + -0.253042 * weathersit_Light Snow & Rain  -0.039913 * weekday_Sunday -0.140478 * windspeed 

#### All the positive coefficients like temp,yr indicate that an increase in these values will lead to an increase in the value of cnt.
#### All the negative coefficients like indicate that an increase in these values will lead to an decrease in the value of cnt.
#### Temp and yr is the most significant with the largest coefficient. which year by year popularity is getting incresed.
#### Followed by weathersit_Light Snow & Rain.
#### This indicates that the bike rentals is majorly affected by temperature,season
